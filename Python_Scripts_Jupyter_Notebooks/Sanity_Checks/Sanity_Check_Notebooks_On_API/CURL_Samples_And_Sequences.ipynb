```python
def check_3_entry_gene(x_gene_array,type_call):
    row_char = []
    for index, row in x_gene_array.iterrows():
        third_char = []
        for i in range(len(row[str(type_call)])):
            
            if (len(row[str(type_call)][i]))==0:
                third_char.append("")
            else:
                third_char.append(row[str(type_call)][i][2])
        row_char.append(set(third_char))
    return row_char
```

```python
def IG_TR_Study(x_gene_array,type_call):
    row_char = []
    for index, row in x_gene_array.iterrows():
        third_char = []
        for i in range(len(row[str(type_call)])):
            
            if (len(row[str(type_call)][i]))==0:
                third_char.append("")
            else:
                if "IG" in row[str(type_call)][i]:
                    third_char.append("IG")
                elif "TR" in row[str(type_call)][i]:
                     third_char.append("TR")
                else:
                    third_char.append(row[str(type_call)][i])
        row_char.append(set(third_char))
    return row_char
```
